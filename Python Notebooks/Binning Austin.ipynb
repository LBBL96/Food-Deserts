```python
import pandas as pd
import numpy as np
from config import gkey2
import requests
import json
```

```python
infile = pd.read_csv("Austin_Distances.csv")
df = pd.DataFrame(infile)
```

```python
df.head()
```

```python
infile = pd.read_csv("Austin_Coords.csv")
df2 = pd.DataFrame(infile)
```

```python
df2.head(2)
```

```python
base_url = "https://maps.googleapis.com/maps/api/distancematrix/json?"
```

```python
school_lat = df2["Lat"]
school_long = df2["Long"]
```

```python
store_address = df["Store Address"]
```

```python
walking_distance = []
i = 0

while i < 60:    
  

    params = {
    "origins": f"{school_lat[i]},{school_long[i]}", 
    "key": gkey2, 
    "destinations": store_address[i], 
    "units": "imperial",
    "mode": "walking"}

    data = requests.get(base_url, params = params).json()
    
    try:

        temp = data["rows"][0]["elements"][0]["distance"]["text"]
        walking_distance.append(float(temp[:-3]))
    except:
        print("not available")
    
    i += 1
```

```python
walking_distance
```

```python
df.count()
```

```python
df["Walking Distance (miles)"] = walking_distance
```

```python
df.to_csv("Walking Austin.csv", index = False)
```

```python
bin_labels = ["< 40", "40-49", "50-59", "60-69", "70-79", "80-89", "90-100"]
bins = [0, 40, 49, 59, 69, 79, 89, 100]
```

```python
df["Poverty Groups"] = df["Percent in Poverty"].copy()
```

```python
df["Poverty Groups"] = pd.cut(df["Poverty Groups"], bins, labels = bin_labels)
```

```python
df.groupby("Poverty Groups")["Driving Distance (miles)"].mean().sort_values(ascending = False)
```
