```python
import pandas as pd
from config import gkey
import gmaps
import numpy as np
import requests

gmaps.configure(api_key=gkey)
```

```python
# Read Laredo data set

Laredo_csv = pd.read_csv("Resources/Laredo_coords.csv")
Laredo_df = pd.DataFrame(Laredo_csv)
Laredo_df.head(19)
```

```python
locations = Laredo_df[["Lat", "Long"]]
locations.head()
```

```python
# Store latitude and longitude in locations variable

locations = Laredo_df[["Lat", "Long"]]

#Fill NaN values and convert to a float

locations = Laredo_df[["Lat", "Long"]].astype(float)
weights = Laredo_df["Percent in Poverty"]

fig = gmaps.figure()

#heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=weights, dissipating=False, max_intensity=100, 
                                point_radius=5)

# add layer
fig.add_layer(heat_layer)

fig
```

```python
# Adding in Hybrid heat map

fig = gmaps.figure(map_type="HYBRID")

#Heat Layer
heat_layer = gmaps.heatmap_layer(locations, weights=weights, dissipating=False, max_intensity=10, point_radius=1)

fig.add_layer(heat_layer)
fig
```
