```python
import geopy.distance as gp
```

```python
closest_store = []
closest_store_index = []
school = 0
school_index = []

i = 0
j = 0
index = 0
closest = 20

while i < len(school_lat):
    
    while j < len(store_lat):
        
        distance = gp.distance(f"{school_lat[i]},{school_long[i]}",f"{store_lat[j]},{store_long[j]}").miles
        
        if distance < closest:
            closest = distance
            index = j
        j += 1                                                              

    closest_store.append(closest)
    closest_store_index.append(index)
    school_index.append(school)
    
    closest = 10
    index = 0
    j = 0
    
    school += 1
    i += 1
```
