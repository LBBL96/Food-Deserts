```python
Laredo_lat = 27.5061111
Laredo_long = -99.5072222

Laredo_grocery = pd.DataFrame()
```

```python
# Setup for google search

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

params = {"key": gkey, "location": f"{Laredo_lat},{Laredo_long}", "radius": 10000, "type": "supermarket"}
```

```python
# Set up URL and perform search
response = requests.get(base_url, params=params)

# convert response to json
places_data = response.json()
```

```python
# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))
```

```python
# Arrays for our necessary info

name = []
lat = []
long = []
```

```python
Laredo_grocery["Store Name"] = name
Laredo_grocery["Location"] = "Laredo"
Laredo_grocery["Lat"] = lat
Laredo_grocery["Long"] = long
```
